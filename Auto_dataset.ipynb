## plot different kinds of graphs for mean value represetation ? 

```python

mean_values = df[['cylinders', 'origin']].mean()


mean_values.plot(kind='bar')
plt.title('Mean Values for Cylinders and Origin')
plt.ylabel('Mean Value')
plt.show()
```

```python
variables = ['mpg', 'displacement', 'weight']

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[variables],)
plt.title('Box Plot of Selected Variables')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

## Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4 ?

```python
sns.pairplot(df[['mpg','weight','origin']],hue='origin',size=4)
```
