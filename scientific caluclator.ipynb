```python
import math

while True:
    print("\n Choose the math Operation's")
    print("0.Addition")
    print("1.Subtraction")
    print("2.Multiplication")
    print("3.Division")
    print("4.Module")
    print("5.Raising Of Power")
    print("6.Square root")
    print("7.Logarithm")
    print("8.Sine")
    print("9.Cosine")
    print("10. Tangent")
    oper = input("\n Your Option from Menu : ")
 
  # ADDITION

    if oper == "0":
        val1 = float(input("\n Enter First Value : ")) 
        val2 = float(input("\n Enter Second Value : "))
        print("\n The Result is : " + str(val1 + val2) + "\n")
        
   # SUBTRACTION

    elif oper == "1":
        val1 = float(input("\n Enter First Value : "))
        val2 = float(input("\n Enter Second Value : "))
        print("\n The Result is : " + str(val1 - val2) + "\n")
        
   # MULTIPLICATION    
    elif oper == "2":
        val1 = float(input("\n Enter First Value : "))
        val2 = float(input("\n Enter Second Value : "))
        print("\n The Result is : " + str(val1 * val2) + "\n")
    
    # DIVISION
    
    elif oper == "3":
        val1 = float(input("\n Enter First Value : "))
        val2 = float(input("\n Enter Second Value : "))
        print("\n The Result is : " + str(val1 / val2) + "\n")
    
    # MODULE
    
    elif oper == "4":
        val1 = float(input("\n Enter First Value : "))
        val2 = float(input("\n Enter Second Value : "))
        print("\n The Result is : " + str(val1 % val2) + "\n")
    
    # RAISING OF POWER
    
    elif oper == "5":
        val1 = float(input("\n Enter First Value : "))
        val2 = float(input("\n Enter Second Value : "))
        print("\n The Result is : " + str(math.pow(val1, val2)) + "\n")
    
    # SQYUARE ROOT
    
    elif oper == "6":
        val1 = float(input("\n Enter Value : "))
        print("\n The Result is : " + str(math.sqrt(val1)) + "\n")
    
    # LOGARITHM
    
    elif oper == "7":
        val1 = float(input("\n Enter First Value : "))
        print("\n The Result is : " + str(math.log(val1)) + "\n")
    
    # SINE
    
    elif oper == "8":
        val1 = float(input("\n Enter First Value : "))
        print("\n The Result is : " + str(math.sin(math.radians(val1))) + "\n")
    
    # COSINE
    
    elif oper == "9":
        val1 = float(input("\n Enter First Value : "))
        print("\n The Result is : " + str(math.cos(math.radians(val1))) + "\n")
    
    # TANGENT
    
    elif oper == "10":
        val1 = float(input("\n Enter First Value : "))
        print("\n The Result is : " + str(math.tan(math.radians(val1))) + "\n")
    
    # # Handling Invalid Data :
    else:
        print("\n Invalid Option")
        continue
        
    back = input("\n Go back to the main menu? (Y/N) : ")
    if back != 'Y':
        break
```
