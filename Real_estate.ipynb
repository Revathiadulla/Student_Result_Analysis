## City wise list all the Villa which is not less than ten thousand

```python
villa_list = data[(data['type'] == 'Villa') & (data['price'] >= 10000)]
print(villa_list)
villa_list_by_city = data.groupby('city',group_keys=False)['street'].apply(list)
print(villa_list_by_city)
```

## In GALT city which residential type has more than 800sq__ft. 


```python
residential_data = data[(data['city'] == 'GALT') & (data['sq__ft'] > 800)]
print(residential_data)
```

## Display their respective details street,sq__ft,sale_date,city.


```python
residential_details=residential_data[['street','sq__ft','sale_date','city']]
residential_details
```

## The cheapest Villa in CA. name the city,street and price for the Villa.

```python

villas_ca = data[(data['type'] == 'villa') & (data['state'] == 'CA')].sort_values(by='price')
cheapest_city = villas_ca['city'].values[0]
cheapest_street = villas_ca['street'].values[0]
cheapest_price = villas_ca['price'].values[0]
print("city : ",cheapest_city)
print("street : ",cheapest_street)
print("price : ",cheapest_price)
```
